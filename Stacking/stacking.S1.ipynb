### Stacking Technique

```python
# importing essential packages

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn import tree
```

```python
# importing dataset

data = pd.read_csv('diabetes.csv')
```

```python
# checking top rows of dataset

data.head()
```

```python
# Creating X and y

X = data.drop('Outcome',axis = 1)
y = data.iloc[:,-1]
```

```python
# As per stacking splitting dataset into 50% training and 50% Hold out data

train,val_train,test,val_test = train_test_split(X,y,test_size=0.5,random_state=123)
```

```python
# Now splitting further Taining dataset into train and test split

x_train,x_test,y_train,y_test = train_test_split(train,test,test_size=0.25,random_state=123)
```

```python
# now on first we try to create KNN classifier

knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
# checking accuracy

knn.score(x_test,y_test)
```

```python
# now training with SVM 

svc = SVC()
svc.fit(x_train,y_train)
```

```python
# Checking accuracy
svc.score(x_test,y_test)
```

```python
# Let's pass Hold out data to both the base model

knn_predict = knn.predict(val_train)
svm_predict = svc.predict(val_train)
```

```python
# now let's see side by side comparision of both model predicted value

predicted_val = np.column_stack((knn_predict,svm_predict))
predicted_val
```

```python
# For test data

predict_test1 = knn.predict(x_test)
predict_test2 = svc.predict(x_test)
```

```python
# now let's see side by side comparision of both model predicted value

predicted_test = np.column_stack((predict_test1,predict_test2))
predicted_test
```

```python
# now we'll create meta model using newly created data

random_forest = RandomForestClassifier()
random_forest.fit(predicted_val,val_test)
```

```python
random_forest.score(predicted_test,y_test)
```
